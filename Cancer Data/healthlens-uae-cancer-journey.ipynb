# 1. Basic Data Understanding and Cleaning

```python
df=pd.read_csv('/kaggle/input/uae-cancer-patient-dataset/_cancer_dataset_uae.csv')
```

```python
df.head(5)
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=['Patient_ID','Death_Date','Cause_of_Death'],inplace=True)
```

```python
for col in {'Gender','Nationality','Emirate','Cancer_Type','Cancer_Stage','Treatment_Type','Hospital','Primary_Physician','Outcome','Smoking_Status','Comorbidities','Ethnicity'}:
    print(df[col].value_counts(),"\n")
```

```python
df.dtypes
```

```python
#convert to datetime
df['Treatment_Start_Date']=pd.to_datetime(df['Treatment_Start_Date'])
#convert to datetime
df['Diagnosis_Date']=pd.to_datetime(df['Diagnosis_Date'])
```

```python
#  Extract month name
df['Treatment_Start_month'] = df['Treatment_Start_Date'].dt.month_name()
df['Diagnosis_Month'] = df['Diagnosis_Date'].dt.month_name()

#  Extract YEAR
df['Treatment_Start_YEAR'] = df['Treatment_Start_Date'].dt.year
df['Diagnosis_Year'] = df['Diagnosis_Date'].dt.year

```

```python
df['Comorbidities'].fillna('None',inplace=True)
```

### C) Time Series Analysis (Trends over Time)

```python
# Count number of patients per year
diagnosis_counts = df['Diagnosis_Year'].value_counts().sort_index()
treatment_counts = df['Treatment_Start_YEAR'].value_counts().sort_index()

treatment=pd.DataFrame(treatment_counts)
diagnosis=pd.DataFrame(diagnosis_counts)

treatment.reset_index(inplace=True)
treatment.columns=["YEAR","Patients_treatment_started"]
diagnosis.reset_index(inplace=True)
diagnosis.columns=["YEAR","Patients_diagnosed"]
```

```python
diagnosis
```

```python
treatment
```

```python
merged=pd.merge(diagnosis,treatment,on='YEAR',how='outer')
merged
```

```python
fig = go.Figure(data=[
    go.Scatter(name='Patients_diagnosed', x=merged['YEAR'], y=merged['Patients_diagnosed'],
               mode='lines+markers', line=dict(color='steelblue')),
    
    go.Scatter(name='Patients_treatment_started', x=merged['YEAR'], y=merged['Patients_treatment_started'],
               mode='lines+markers', line=dict(color='lightcoral'))
])

fig.update_layout(
    title='Diagnosed vs Treatment Started Per Year',
    xaxis_title='Year',
    yaxis_title='Number of Patients',template='plotly_dark',
    xaxis=dict(
        tickmode='array',
        tickvals=['2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']
    )
)

fig.show(renderer='iframe')
```

```python
# Inference:
# This line chart compares the number of patients diagnosed and those who started treatment year-wise.
# Any gap between diagnosis and treatment lines may indicate delay or lack of treatment availability.
# Also shows temporal trends and possible impact of events like COVID-19.
```

```python
fig = px.histogram(df, x='Treatment_Start_YEAR', color='Outcome', 
                   color_discrete_sequence=px.colors.sequential.Sunset,
                   category_orders={
                 'Treatment_Start_YEAR': ['2015','2016','2017','2018', '2019', '2020', '2021', '2022','2023','2024'],  
                 'Outcome': ['Recovered', 'Under Treatment', 'Deceased'] },barmode='stack',
                   title='Outcome vs Treatment Start Year')
fig.update_layout(bargap=0.4,template='plotly_dark',
    xaxis=dict(
        tickmode='array',
        tickvals=['2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']
    )
)
fig.update_traces(marker_line_width=0.1, marker_line_color='black')
fig.show(renderer='iframe')
```

###  F) BMI vs Outcome

```python
df['BMI']=df['Weight']/(df['Height']/100)**2
```

```python

corr_cols = ['Age', 'Weight', 'Height', 'BMI']
corr = df[corr_cols].corr().round(2)

fig = go.Figure(data=go.Heatmap(
    z=corr.values,
    x=corr.columns,
    y=corr.columns,
    colorscale='Cividis',
    text=corr.values,
    texttemplate="%{text}"
))
fig.update_layout(title='Correlation Heatmap',template='plotly_dark')
fig.show(renderer='iframe')
```

```python
# BMI Distribution
fig = px.histogram(df, x='BMI', nbins=30, marginal='box', title='BMI Distribution')
fig.update_layout(
    xaxis_title='BMI',
    yaxis_title='Number of Patients',
    template='plotly_dark'
)
fig.update_traces(marker_line_width=0.25, marker_line_color='black')
fig.show(renderer='iframe')
```

```python
# BMI vs Outcome
fig = px.box(df, x='Outcome', y='BMI', color='Outcome', title='BMI vs Outcome')
fig.update_layout(
    template='plotly_dark'
)
fig.show(renderer='iframe')
```

```python
# Shows the distribution of BMI among patients and includes a boxplot for spread.
# Helps understand the general weight profile and identify outliers.
# Useful in correlating obesity with cancer prevalence or outcomes.
```
